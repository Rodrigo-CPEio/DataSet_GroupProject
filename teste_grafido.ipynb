```python
import pandas as pd
import plotly.express as px

# Cargar dataset
dataset = pd.read_csv("./DATASET/dataset_limpo_medianag.csv")

# Filtrar salarios válidos (1.000 a 500.000)
df_plot = dataset[
    (dataset['ConvertedCompYearly'] > 1000) &
    (dataset['ConvertedCompYearly'] < 500000)
].copy()

# Crear boxplot interactivo con Plotly
fig = px.box(
    df_plot,
    x="Geração",                 # Eje X: generación
    y="ConvertedCompYearly",      # Eje Y: salario
    color="Geração",              # Color por generación
    labels={"ConvertedCompYearly": "Salário Anual (USD)", "Geração": "Geração"},

)

# Escala logarítmica en Y (essencial para datos salariales)
fig.update_yaxes(type="log")

# Ajustes de layout opcionales
fig.update_layout(
    title_font_size=18,
    xaxis_title_font_size=14,
    yaxis_title_font_size=14,
    margin=dict(l=20, r=20, t=50, b=20),
)

# Guardar como HTML para incrustar en dashboard
fig.write_html(
    "C:/Users/andre/OneDrive/Ambiente de Trabalho/dashoard_UPT/DataSet_GroupProject/Public/assets/boxplot_geracao.html"
)
```
